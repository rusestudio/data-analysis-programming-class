```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.datasets import fetch_openml
from sklearn.utils import shuffle
import pandas as pd
```

```python
#load mnist data
mnist = fetch_openml('mnist_784')
```

```python
#label 
X, y = mnist["data"], mnist["target"]
```

```python
X
```

```python
y
```

```python
#choose random data sample 1000
#X,y = shuffle(X,y, random_state=42)
X_sample , y_sample = X[:1000], y[:1000]
```

```python
y_sample.astype
```

```python
#use t-sne
#n_compement mean how many dimenseion= 2 = 2d
#add random_state to get same data and not random data
tsne = TSNE(n_components=2) #but here not add random_state
X_reduced = tsne.fit_transform(X_sample)
```

```python
X_reduced
X_reduced.shape
```

```python
# make scatter plot use tsne

plt.figure(figsize=(10,8)) #make canvad
scatter = plt.scatter(X_reduced[:,0], 
                      X_reduced[:,1],
                      c=y_sample.astype(int), #mapping number for each color
                      cmap="tab10", #10 color tab for 10 number
                      alpha=0.7 # transparency
                      ) 

plt.colorbar(scatter, ticks=range(10))
plt.title('tsne visualization of 1000 Mnist sample')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show
```
