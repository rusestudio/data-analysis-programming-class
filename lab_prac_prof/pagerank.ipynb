```python

#make graph edge
G = nx.DiGraph()
G.add_edge("A", "B")
G.add_edge("B", "D")
G.add_edge("C", "A")
G.add_edge("C", "B")
G.add_edge("C", "D")
G.add_edge("D", "C")
G.add_node("E")
```

```python

#calc page rank
page_rank = nx.pagerank(G, alpha=0.85) #alpha is damping factor default value/prob
# 85% follow link iter, 15% random
```

```python
import networkx as nx

# 실제 데이터 로드 - Karate Club 네트워크 사용
#use karte club network graph
G = nx.karate_club_graph()

# PageRank 계산
#calc page rank score damping factor 0.85, global importance, centralily
pagerank_scores = nx.pagerank(G, alpha=0.85)  # alpha는 damping factor

# 결과 출력
print("PageRank Scores for Karate Club Network:")
#print each node score
for node, score in pagerank_scores.items():
    print(f"Node {node}: {score:.4f}")
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# 그래프 시각화
plt.figure(figsize=(10, 7))

# 노드 크기 설정 (PageRank 점수에 비례)
#set node size based on pagerank score node not real but 비율
node_sizes = [pagerank_scores[node] * 3000 for node in G.nodes()]

# 노드 시각화
nx.draw(
    G,
    pos=nx.spring_layout(G),  # 레이아웃 설정 (spring layout 사용)
    with_labels=True,
    node_size=node_sizes,  # 노드 크기 설정
    node_color="skyblue",
    edge_color="gray",
    font_size=10,
    font_weight="bold",
)

plt.show()
```

```python
#hyperlink induced topic search . calc hits score.- link structure
hubs, authorities = nx.hits(G) 

#node is good hub if point to many good authority
#how good node link to important node
print("Hub Scores:")
for node, score in hubs.items(): #calc hub item score for each node
    print(f"Node {node}: {score:.4f}") 

#node is good autority if many hub point to it. 
#how reliable this node are
print("\nAuthority Scores:")
for node, score in authorities.items(): #calc autho item score for each node
    print(f"Node {node}: {score:.4f}") 
```
