```python
import numpy as np

from sklearn.datasets import make_blobs

#from data set take 2000 sample data
n_samples = 2000

#make 3 array with 3 center, 3 cluster. 
X, _ = make_blobs(
    n_samples=n_samples,
    centers=np.array([[0, -2.5], [2.5, 2.5], [-2.5, 2.5]]),
    cluster_std=1.0,
)
```

```python
import matplotlib.pyplot as plt

#use scatter to plot current array
fig = plt.figure(figsize=(12, 8))
plt.scatter(X[:, 0], X[:, 1])
fig.show()
```

```python
from sklearn.cluster import KMeans

#make 3 cenroid = 3 cluster
kmeans = KMeans(n_clusters=3)
kmeans.fit(X) #train data to cluster based on 3 centroid to make 3 cluster

kmeans_predicted = kmeans.predict(X) #predict kmean cluster
```

```python
fig = plt.figure(figsize=(12, 8))
plt.scatter(X[:, 0], X[:, 1], c=kmeans_predicted) #use center from predict kmean

#make kmean as cluster centers
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], c="red", marker="x", s=100) #make centroid x 
fig.show()
```

```python
from sklearn.cluster import DBSCAN

#density based clustering
#identify core, border, noise, based on epilson distance
dbscan = DBSCAN(eps=0.5, min_samples=8) #epilson 0.5 , min_samples- min number data to consider dense
dbscan_predicted = dbscan.fit_predict(X) # perform clustering and return cluster label
```

```python
fig = plt.figure(figsize=(12, 8))
plt.subplot(111)

#make cluster based on predicted
plt.scatter(X[:, 0], X[:, 1], c=dbscan_predicted)

#make labels based dbscan labels
labels = dbscan.labels_

unique_labels = set(labels)
colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))


fig.show()
```
