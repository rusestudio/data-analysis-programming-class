```python
from kiwipiepy import Kiwi
from kiwipiepy.utils import Stopwords

kiwi = Kiwi(model_type="sbg")
stopwords = Stopwords() #add stop word so non meaningful word will not go into the visualization
```

```python
kiwi.add_user_word("다야나", "NNp") #put name in dic so that it will not get cut at middle
kiwi.tokenize("안녕하세요 제 이름은 다야나입니다")
```

```python
from collections import Counter
from tqdm import tqdm

analyzing_tags = {"NNG", "NNP", "VV", "VA", "SL"}
counter = Counter()


for news in tqdm(data, mininterval=1):
    text = news["text"].strip() #strip space in text
    if len(text) < 10:
        continue

    tokens = [
        tokenized.form for tokenized in  kiwi.tokenize(text, stopwords=stopwords)
        if tokenized.tag in analyzing_tags and len(tokenized.form) > 1
        ]

    counter.update(tokens)
    

```

```python
for token, freq in counter.most_common(30):
    print(token, freq)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

most_freq = counter.most_common(100)

wordcloud = WordCloud(
    font_path='C:\\Users\\user\\AppData\\Local\\Microsoft\\Windows\\Fonts\\NanumGothic.ttf',
    prefer_horizontal = 1.0,
    background_color = "white",
)

cloud = wordcloud.generate_from_frequencies(dict(most_freq)) #dict type , so need to change tuple-> dict

plt.figure(figsize=(10,8))
plt.imshow(cloud)
plt.show()
```
