```python
class r:
    def __init__(self, start,end):
        self.start = start
        self.end = end

    def __iter__(self):
        return self
    
    def __next__(self):
        if self.start >= self.end:
            raise StopIteration
        
        value = self.start
        self.start += 1 #to make go to next value after start
        return value
    
r1 = r(1,10)

num = 0
for i in r1:
    num += i

print(num)
```
