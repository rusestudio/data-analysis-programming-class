```python
import torch
from torch import nn
from torch.nn import functional as F
from typing import List

class MLPClassifier(nn.Module):
    def __init__(self, input_dimension: int, hidden_layer_sizes: List[int]):
        super(MLPClassifier).__init__()
        self.input_dimension = input_dimension
        self.hidden_layers = nn.ModuleList()
        previous_size = input_dimension
        for hiden_size in hidden_layer_sizes:
            self.hidden_layers.append(nn.Linear(previous_size, hiden_size)) #edge between layers
            previous_size = hiden_size

        self.output_linear = nn.Linear(previous_size, 10)

    def forward(self, x: torch.Tensor) -> torch.Tensor: # to make foword
        x  = x.view(-1, self.hidden_layers)
        for layer in self.hidden_layers:
            x = F.gelu(layer(x)) #activication layer gelu    

        z_3 = self.output_linear(x)
        return z_3
```
