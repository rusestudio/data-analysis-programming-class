```python
import networkx as nx
```

```python
G = nx.DiGraph()
G.add_edge("A", "B")
G.add_edge("B", "D")
G.add_edge("C", "A")
G.add_edge("C", "B")
G.add_edge("C", "D")
G.add_edge("D", "C")
G.add_node("E")
```

```python
page_rank = nx.pagerank(G, alpha=0.85) #alpha= damping factor d, random
```

```python
import networkx as nx
```

```python
G = nx.karate_club_graph() #example graph from karate
```

```python
print(G)
```

```python
pagerank_score = nx.pagerank(G, alpha=0.85)

for node, score in pagerank_score.items():
    print(f"node:{node}, {score}" )
```

```python
import  matplotlib.pyplot as plt

figure = plt.figure(figsize=(10,8))

node_sizes = [pagerank_score[node]* 9000 for node in G.nodes()]

nx.draw(
    G,
    pos=nx.spring_layout(G), #position. act like spring. if node near push if node far pull like spring extend
    with_labels=True,
    node_size= node_sizes,
    node_color="skyblue",
    edge_color="gray",
    font_size=11,
    font_weight="bold",

    
)
```
