```python
import numpy as np
from sklearn.datasets import make_blobs

num_samples = 2000

x, _ = make_blobs(
    n_samples=num_samples,
    n_features=2,
    centers= np.array([[0 , -2,5], [2.5, 2.5] , [-2.5, 2.5]]),
    cluster_std=1.0
    )
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12,8))
plt.scatter(x[:,0], x[:,1])
plt.show()
```

```python
from sklearn.cluster import KMeans

kmean = KMeans(n_clusters=3)

kmean.fit(x)

predicted = kmean.predict(x)
```

```python
predicted.shape
```

```python
fig =plt.figure(figsize=(12,8))
plt.scatter(x[:,0], x[:,1], c=predicted)

centriod = kmean.cluster_centers_
plt.scatter(centriod[:,0], centriod[:,1], c="red", marker="x" , s=100)

fig.show()
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN( eps= 0.5, min_samples=8)
predict= dbscan.fit_predict(x)
```

```python
predict[:100] #-1 = noise
```

```python
label= dbscan.labels_
uni_label = set(label)
```

```python
uni_label #-1= noide 0= cluster 1 1= cluster 2
```
